####  1. Compute min, 25th, median, 75th, and max of a Series ser

```python
import pandas as pd
import numpy as np

ser = pd.Series(np.random.randint(1, 100, 1000))  # Example series
summary = ser.describe(percentiles=[.25, .5, .75])
print(summary[['min', '25%', '50%', '75%', 'max']])
```

#### 9. Create a Line Plot

```python
import matplotlib.pyplot as plt

x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

plt.plot(x, y, marker='o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot of X vs Y")
plt.grid(True)
plt.show()
```

#### 10 Plot x, x², x³, x⁴ on same plot

```python
x = np.arange(0, 6.1, 0.3)

plt.plot(x, x, 'ro', label='x')       # red circles
plt.plot(x, x**2, 'bs', label='x²')   # blue squares
plt.plot(x, x**3, 'g', label='x³')    # green line
plt.plot(x, x**4, ':', label='x⁴')    # dotted line

plt.xlim(0, 6)
plt.ylim(0, 125)
plt.title('Powers of x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

#### 11 Bar Plot for Heights

```python
height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

plt.bar(names, height, color='skyblue')
plt.title("Height Comparison")
plt.xlabel("Individuals")
plt.ylabel("Height (cm)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 12: Histogram with Normal Distribution

```python
x = np.random.randn(100000)

# 10 bins
plt.hist(x, bins=10, color='steelblue', edgecolor='black')
plt.title("Histogram with 10 bins")
plt.show()

# 20 bins
plt.hist(x, bins=20, color='lightgreen', edgecolor='black')
plt.title("Histogram with 20 bins")
plt.show()

# 50 bins
plt.hist(x, bins=50, color='salmon', edgecolor='black')
plt.title("Histogram with 50 bins")
plt.show()
```

#### Q13 Features of TensorFlow
Open-source deep learning library by Google.

Works on CPU, GPU, and TPU.

Supports automatic differentiation and gradient computation.

Highly scalable for distributed computing.

Has high-level APIs like Keras.

Good for both training and deployment (mobile, web, edge).

Visualization via TensorBoard.

Supports TensorFlow Lite for mobile apps and TensorFlow Serving for production.

#### Q14 Limitations of TensorFlow

Steeper learning curve compared to PyTorch (before TF 2.x).

Verbosity in code (especially in early versions).

Debugging can be harder than eager-mode frameworks.

Performance optimization requires experience.

Sometimes slow for small models or datasets.

Errors may not always be user-friendly.

#### Q15: Supervised vs Unsupervised Machine Learning


| Aspect       | Supervised Learning                            | Unsupervised Learning                         |
|--------------|------------------------------------------------|-----------------------------------------------|
| Data         | Uses labeled data (input + output)             | Uses only input data (no labels)              |
| Goal         | Predict output or classify data                | Discover patterns or groupings                |
| Examples     | Regression, Classification                     | Clustering, Dimensionality Reduction          |
| Algorithms   | Linear Regression, SVM, Decision Trees         | K-Means, PCA, Hierarchical Clustering         |
| Use Case     | Email spam detection, stock prediction         | Customer segmentation, anomaly detection      |
