####  1. Compute min, 25th, median, 75th, and max of a Series ser

```python
import pandas as pd
import numpy as np

ser = pd.Series([12, 7, 3, 18, 25, 6, 15, 10])

min_val = ser.min()
q1 = ser.quantile(0.25)
median = ser.median()
q3 = ser.quantile(0.75)
max_val = ser.max()

print("Minimum:", min_val)
print("25th percentile (Q1):", q1)
print("Median (Q2):", median)
print("75th percentile (Q3):", q3)
print("Maximum:", max_val)
```

#### 10 Plot x, x², x³, x⁴ on same plot

```python
x = np.arange(0, 6.1, 0.3)

plt.plot(x, x,'ro', label='x')       
plt.plot(x, x**2, 'bs', label='x²')   
plt.plot(x, x**3, 'g', label='x³')    
plt.plot(x, x**4, ':', label='x⁴')    

plt.xlim(0, 6)
plt.ylim(0, 125)
plt.title('Powers of x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(axis='y')
plt.show()
```

#### 11 Bar Plot for Heights

```python
height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

plt.bar(names, height, color='skyblue')
plt.title("Height Comparison")
plt.xlabel("Individuals")
plt.ylabel("Height (cm)")
plt.grid(axis='y', linestyle='--', alpha=.4)
plt.show()
```

#### 12: Histogram with Normal Distribution

```python
x = np.random.randn(100000)

plt.hist(x, bins=10, color='steelblue', edgecolor='black')
plt.title("Histogram with 10 bins")
plt.show()

plt.hist(x, bins=20, color='lightgreen', edgecolor='black')
plt.title("Histogram with 20 bins")
plt.show()

plt.hist(x, bins=50, color='salmon', edgecolor='black')
plt.title("Histogram with 50 bins")
plt.show()
```

#### Q15: Supervised vs Unsupervised Machine Learning


| Aspect       | Supervised Learning                            | Unsupervised Learning                         |
|--------------|------------------------------------------------|-----------------------------------------------|
| Data         | Uses labeled data (input + output)             | Uses only input data (no labels)              |
| Goal         | Predict output or classify data                | Discover patterns or groupings                |
| Examples     | Regression, Classification                     | Clustering, Dimensionality Reduction          |
| Algorithms   | Linear Regression, SVM, Decision Trees         | K-Means, PCA, Hierarchical Clustering         |
| Use Case     | Email spam detection, stock prediction         | Customer segmentation                         |
